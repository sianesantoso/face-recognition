```python
import os
from PIL import Image
```

```python
folder_dir = "augmented_images"  
jpeg_folder = "gambar_jpeg/augmented_images"  # Folder tujuan untuk menyimpan file jpeg
```

```python
# Membuat folder tujuan jika belum ada
os.makedirs(jpeg_folder, exist_ok=True)
```

```python
# Fungsi untuk konversi file .ppm ke .jpeg
def convert_ppm_to_jpeg(ppm_path, jpeg_path):
    with Image.open(ppm_path) as img:
        img.convert("RGB").save(jpeg_path, "JPEG")
```

```python
for root, _, files in os.walk(folder_dir):
    for file in files:
        if file.endswith(".ppm"):
            ppm_path = os.path.join(root, file)

            # Membuat path tujuan dengan struktur yang sama
            relative_path = os.path.relpath(root, folder_dir)
            target_folder = os.path.join(jpeg_folder, relative_path)
            os.makedirs(target_folder, exist_ok=True)

            # Tentukan path file jpeg baru
            jpeg_path = os.path.join(target_folder, file.replace(".ppm", ".jpeg"))

            # Konversi dan simpan ke folder tujuan
            convert_ppm_to_jpeg(ppm_path, jpeg_path)
            print(f"Converted: {ppm_path} -> {jpeg_path}")

print("Konversi selesai.")
```
